# Mars Weather Data

> This is data representing the weather conditions on Mars from Sol 1 (August 7, 2012 on Earth) to Sol 1895 (February 27, 2018 on Earth).

>Source(s) & Methodology: This data was measured and transmitted via the Rover Environmental Monitoring Station (REMS) on-board the Curiosity Rover. The data was made publicly available by NASA’s Mars Science Laboratory and the Centro de Astrobiología (CSIC-INTA).

```python
from matplotlib import pyplot as plt 
import numpy as np

plt.rcdefaults() # set the matplotlib runtime configuration (rc) to its defaults

# If you're using Jupyter Notebook, you may also want to run the following
# line of code to display your code in the notebook:
%matplotlib inline 
```

```python
# This gives us access to the library that will read a csv file:

import csv 
```

```python
with open('mars-weather.csv') as file:
    table = [row for row in csv.reader(file)]
```

```python
# This will tell me if a value is convertible to a float
def is_float(s):
    try:
        f = float(s)
    except:
        return False
    return True
```

### Getting the column with sols

```python
## The row with sols as strings

sols_str = [table[i][2] for i in range(len(table)) if (i != 0 )]
```

```python
## Using numpy to convert sols_str into a numpy int32 array
### I also flip the whole thing

sols = np.flipud(np.array(sols_str, dtype="int32"))
sols
```

### Getting the column with pressure

```python
## The row with pressure as strings

press_str = [table[i][-3] for i in range(len(table)) if (i != 0 )]
```

```python
press_str
```

```python
## Getting all pressure values into a numpy array as floats
### I also reverse the array

press = np.flipud(np.array(press_str, dtype="float32"))
```

```python
press
```

## A Really interesting fact
#### 1 Martian Year = 668 sols, which is where the periodicity of the graph below comes from.
#### If you pay attention, the period of the cycle ends up being ~650 sols, according to the graph below

*Some pressure data is just not present (NaN) leading to some gaps in the graph*

```python
plt.figure(figsize=(10,8))
plt.ylabel('Pressure in mb')
plt.xlabel('Sols')
plt.title("Pressure on the surface of Mars vs. Sols (Martian days)")
plt.plot(sols, press, 'purple')
plt.show()
```
